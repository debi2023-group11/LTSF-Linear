```python
import subprocess

# Define the values for preLen
preLens = [24, 36, 48, 60]

# Define the common arguments for the run.py command
common_args = [
    "python",
    "-u",
    "run.py",
    "--is_training",
    "1",
    "--root_path",
    "./dataset/",
    "--data_path",
    "air_quality_ts.csv",
    "--task_id",
    "air",
    "--model",
    "FEDformer",
    "--data",
    "custom",
    "--features",
    "S",
    "--seq_len",
    "36",
    "--label_len",
    "18",
    "--e_layers",
    "2",
    "--d_layers",
    "1",
    "--factor",
    "3",
    "--enc_in",
    "1",
    "--dec_in",
    "1",
    "--c_out",
    "1",
    "--des",
    "Exp",
    "--itr",
    "3"
]

# Loop through each preLen and execute the run.py command with the corresponding pred_len value
for preLen in preLens:
    args = common_args + ["--pred_len", str(preLen)]
    subprocess.run(args)
```
